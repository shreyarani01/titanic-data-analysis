```python
#group by pclass then find: survival rate, avg. fare, avg. age
```

```python
df.groupby("Pclass").size() #number of passengers in each Pclass
```

```python
df.groupby("Pclass")["Survived"].mean() #survival rate in each pclass
```

```python
df.groupby("Pclass")["Fare"].mean() #average fare in each pclass
```

```python
df.groupby("Pclass")["Age"].mean() #average age in each pclass
```

```python
df["Survived"].value_counts() # 0 means not survived, 1 means survived
```
